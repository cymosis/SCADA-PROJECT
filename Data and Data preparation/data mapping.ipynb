```python

# Convert ISO 8601 timestamps to datetime, ignore invalid formats
attack_data['GMT +0'] = pd.to_datetime(
    attack_data['GMT +0'],
    errors='coerce',           # ignore invalid timestamps
    infer_datetime_format=True
)

# Convert from UTC (Z) to GMT+8
attack_data['GMT +8'] = attack_data['GMT +0'].dt.tz_convert('Etc/GMT-8')  # tz-aware

# Optional: 12-hour format with AM/PM
attack_data['GMT_12hr'] = attack_data['GMT +8'].dt.strftime('%Y-%m-%d %I:%M:%S %p')

# Check results
attack_data.head()
```

```python

attack_data = attack_data.dropna(subset=['GMT_12hr'])

# Convert 12-hour string back to datetime (no timezone)
attack_data['GMT_12hr_dt'] = pd.to_datetime(
    attack_data['GMT_12hr'])
attack_data = attack_data.dropna(subset=['GMT_12hr'])

# Check results
attack_data.head()
```

```python
attack_data['GMT_12hr_dt'].min(),attack_data['GMT_12hr_dt'].max()
```

```python
attack_windows = [
    ('2019-07-20 03:08:46 PM', '2019-07-20 03:10:31 PM'),  # Attack on FIT401
    ('2019-07-20 03:15:00 PM', '2019-07-20 03:19:32 PM'),  # Attack on LIT301
    ('2019-07-20 03:26:57 PM', '2019-07-20 03:30:48 PM'),  # Attack on P601
    ('2019-07-20 03:38:50 PM', '2019-07-20 03:46:20 PM'),  # Multi-point Attack
    ('2019-07-20 03:54:00 PM', '2019-07-20 03:56:00 PM'),  # Attack on MV501
    ('2019-07-20 04:02:56 PM', '2019-07-20 04:16:18 PM')   # Attack on P301
]
```

```python


# Flag attacks
attack_data['Attack'] = 0
for start, end in attack_windows:
    attack_data.loc[(attack_data['GMT_12hr_dt'] >= start) & 
                    (attack_data['GMT_12hr_dt'] <= end), 'Attack'] = 1
```

```python
num_attacks = attack_data['Attack'].sum()
print(f"Number of attack rows: {num_attacks}")
```

```python
labelled_attack_data = attack_data.drop(columns=['GMT +8', 'GMT_12hr', 'GMT_12hr_dt'])
labelled_attack_data.head()
labelled_attack_data.to_excel('SWaT_Attack_Data_Labeled.xlsx', index=False)
```

```python
labelled_attack_data=labelled_attack_data.rename(columns={'GMT +0':'t_stamp'})
labelled_attack_data.head()
```

```python
# Convert all timezone-aware datetime columns to timezone-naive
for col in labelled_attack_data.select_dtypes(include=['datetimetz']).columns:
    labelled_attack_data[col] = labelled_attack_data[col].dt.tz_localize(None)

# Save to Excel
labelled_attack_data.to_excel('SWaT_Attack_Data_Labeled.xlsx', index=False)
```
